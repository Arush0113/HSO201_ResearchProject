```
import pandas as pd
df = pd.read_csv('/content/table1_hso(2).txt', delimiter=' ')
```

```
df2 = df.drop(columns=['Fall_2018','Fall_2019','Fall_2020','Fall_2021','Fall_2022','Fall_2029','Percentage_change_enrollment_2012_to_2017','Percentage_change_enrollment_2017_to_2029'], axis=1)
df2 = df2[df2['Regions'] != '1']
df2 = df2[df2['Regions'] != 'State']
df2 = df2[df2['Regions'] != 'Region']
df2
```

```
# df2.drop(['0','2','7'], axis = 1)
df3 = pd.DataFrame.transpose(df2)
df3.columns = df3.iloc[0]
df4 = df3.drop(["Regions"])
df4
```

```
for col in df4.columns:
  l = []
  for items in df4[col]:
    l.append(items.replace(',',''))
  df4[col] = l
  df4[col].astype(str).astype(int)
  print(df4[col])
```

```
from bokeh.palettes import Spectral8
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

date = [ 	1990 ,	2000, 	2007 ,	2008, 	2009 ,	2010 ,	2011 ,	2012 ,	2013 ,	2014 ,	2015 ,	2016 ,	2017]

df_list = ['Alabama', 	'Alaska' ,	'Arizona' ,	'Arkansas'  ,	'Colorado' ,	'Connecticut' ,	'Delaware' ,	'District_of_Columbia' ,	'Florida' ,	'Georgia' ,	'Hawaii' ,	'Idaho' ,	'Illinois' ,	'Indiana' ,	'Iowa' ,	'Kansas' ,	'Kentucky', 	'Louisiana' ,	'Maine' ,	'Maryland' ,	'Massachusetts' ,	'Michigan' ,	'Minnesota' ,	'Mississippi' ,	'Missouri','Montana' 	,'Nebraska' ,	'Nevada' ,	'New_Hampshire' ,	'New_Jersey' ,	'New_Mexico' 	,'New_York' ,	'North_Carolina' ,	'North_Dakota' ,	'Ohio' ,	'Oklahoma' ,	'Oregon' ,	'Pennsylvania' ,	'Rhode_Island' ,	'South_Carolina' ,	'South_Dakota' ,	'Tennessee' ,	'Texas' ,	'Utah' ,	'Vermont' ,	'Virginia' ,	'Washington', 	'West_Virginia' ,	'Wisconsin' ,	'Wyoming']
dfl = [df4[i] for i in df_list]

p = figure(plot_width=1000, plot_height=550)
p.title.text = 'Click on legend entries to mute the corresponding lines'

for data, name, color in zip(dfl, df_list, Spectral8):
    p.line(date, data, line_width=2, color=color, alpha=0.8,
          muted_color=color, muted_alpha=0.2, legend_label=name)
    

p.legend.location = "top_left"
p.legend.click_policy="mute"

show(p)
```
